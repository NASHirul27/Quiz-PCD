```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/avatar-the-last-airbender-3.png', 0)
```

```python
cv2_imshow(img)
```

```python
height, width = img.shape

print("Lebar gambar = ", width)
print("Tinggi gambar = ", height)
```

```python
min_val, max_val, _, _ = cv2.minMaxLoc(img)

print("Nilai Minimum Pixel:", min_val)
print("Nilai Maksimum Pixel:", max_val)
```

```python
normalized_img = img / 255.0
print("Nilai Piksel Terormalisasi:")
print(normalized_img)
```

```python
flipped_horizontal = cv2.flip(img, 1)
cv2_imshow(flipped_horizontal)
```

```python
flipped_vertical = cv2.flip(img, 0)
cv2_imshow(flipped_vertical)
```

```python
import numpy as np
```

```python
# Tentukan parameter Gaussian noise
mean = 0
std_dev = 25

# Buat noise Gaussian
gaussian_noise = np.random.normal(mean, std_dev, img.shape).astype('uint8')

# Tambahkan noise ke gambar
gaussian_noisy_image = cv2.add(img, gaussian_noise)

# Menampilkan gambar asli dan gambar dengan noise Gaussian
cv2_imshow(gaussian_noisy_image)
```

```python
# Tentukan probabilitas untuk noise salt & pepper
salt_pepper_prob = 0.05

# Buat mask salt & pepper
salt_pepper_mask = np.random.choice([0, 1, 2], size=img.shape[:2], p=[
                                     1 - salt_pepper_prob, salt_pepper_prob / 2., salt_pepper_prob / 2.]).astype('uint8')

# Buat noise salt & pepper
salt_noise = (salt_pepper_mask == 1) * 255
pepper_noise = (salt_pepper_mask == 2) * 0

# Tambahkan noise ke gambar
salt_noisy_image = cv2.add(cv2.add(img, salt_noise), pepper_noise)

# Menampilkan gambar asli dan gambar dengan noise salt & pepper
cv2_imshow(salt_noisy_image)
```

```python
import random
# Ukuran crop
crop_height, crop_width = 250, 250

# Ambil ukuran gambar
height, width = img.shape[:2]

# Tentukan koordinat awal untuk crop secara acak
start_x = random.randint(0, width - crop_width)
start_y = random.randint(0, height - crop_height)

# Lakukan crop
cropped_image = img[start_y:start_y+crop_height, start_x:start_x+crop_width]

# Menampilkan gambar asli dan hasil crop secara acak
cv2_imshow(cropped_image)
```

```python
# Ukuran crop
crop_height, crop_width = 250, 250

# Ambil ukuran gambar
height, width = img.shape[:2]

# Tentukan koordinat awal untuk crop dari pusat gambar
start_x = (width - crop_width) // 2
start_y = (height - crop_height) // 2

# Lakukan crop
cropped_image = img[start_y:start_y+crop_height, start_x:start_x+crop_width]

# Menampilkan gambar asli dan hasil center crop
cv2_imshow(cropped_image)
```
